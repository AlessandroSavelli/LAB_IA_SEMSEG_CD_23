```python
#CREAZIONE DATASET SENZA NUVOLE ---> IMMAGINI

import os
import shutil
import pandas as pd

# Specifica la directory delle immagini
image_directory = "/content/drive/MyDrive/Colab Notebooks/sentinel2png_da_usare"

# Carica il file CSV come elenco di righe
csv_file = "/content/drive/MyDrive/Colab Notebooks/Train_s2_clouds_ORDINATO.csv"  # Sostituisci con il nome del tuo file CSV
with open(csv_file, 'r') as file:
    csv_data = file.readlines()

# Crea una nuova directory per le immagini con valore 0 nella terza colonna
output_directory = "/content/drive/MyDrive/Colab Notebooks/immagini_senza_nuvole"
os.makedirs(output_directory, exist_ok=True)

# Lista per i percorsi delle immagini ordinate
sorted_image_paths = []

# Ottieni prima l'ordine alfabetico delle sottocartelle
subdirs = sorted([subdir for subdir in os.listdir(image_directory) if os.path.isdir(os.path.join(image_directory, subdir))])

# Itera sulle sottocartelle ordinate
for subdir in subdirs:
    subdir_path = os.path.join(image_directory, subdir)

    # Ottieni l'ordine alfabetico dei file all'interno di ciascuna sottocartella
    files = sorted([file for file in os.listdir(subdir_path) if os.path.isfile(os.path.join(subdir_path, file))])

    # Aggiungi i percorsi delle immagini alla lista
    for filename in files:
        image_path = os.path.join(subdir_path, filename)
        sorted_image_paths.append(image_path)

immagini_copiate = 0

# Itera sulle righe del file CSV e sugli elementi nell'elenco delle immagini
for csv_row, image_path in zip(csv_data, sorted_image_paths):
    # Estrai il valore dalla terza colonna nel file CSV
   try:
        value = int(csv_row.strip().split(',')[2])
   except ValueError:
        value = -1  # Se c' un errore, assegna un valore che non verr copiato

   if value == 0:
        # Copia l'immagine nella nuova directory
        destination_path = os.path.join(output_directory, os.path.relpath(image_path, image_directory))
        os.makedirs(os.path.dirname(destination_path), exist_ok=True)
        try:
            shutil.copy2(image_path, destination_path)
            immagini_copiate += 1
            # Stampa il percorso dell'immagine copiata e il numero di immagini copiate
            print(f"Copia ({immagini_copiate} immagini copiate): {image_path}")
        except Exception as e:
            print(f"Errore durante la copia di {image_path}: {str(e)}")

print(f"Totale {immagini_copiate} immagini con valore 0 nella terza colonna copiate nella nuova directory con le sottocartelle e ordinamento specifico.")
```

```python
import os

# Funzione ricorsiva per contare le foto in una cartella e nelle sottocartelle
def conta_foto(cartella):
    # Inizializza una variabile per contare le foto
    numero_di_foto = 0

    # Estensioni di file comuni per le foto
    estensioni_foto = [".jpg", ".jpeg", ".png", ".gif", ".bmp", ".tiff"]

    # Itera attraverso tutti gli elementi nella cartella
    for elemento in os.listdir(cartella):
        percorso_elemento = os.path.join(cartella, elemento)

        # Se l'elemento  una cartella, richiama la funzione ricorsivamente
        if os.path.isdir(percorso_elemento):
            numero_di_foto += conta_foto(percorso_elemento)
        # Se l'elemento  un file e ha un'estensione di foto
        elif os.path.isfile(percorso_elemento) and any(percorso_elemento.endswith(ext) for ext in estensioni_foto):
            numero_di_foto += 1

    return numero_di_foto

# Definisci la cartella principale
cartella_principale = "/content/drive/MyDrive/Colab Notebooks/maschere_senza_nuvole"

# Ottieni una lista ordinata delle sottocartelle in ordine alfabetico
sottocartelle_ordinate = sorted([sottocartella for sottocartella in os.listdir(cartella_principale) if os.path.isdir(os.path.join(cartella_principale, sottocartella))])

# Inizializza una variabile per il conteggio totale di foto
totale_foto = 0

# Itera attraverso le sottocartelle ordinate
for sottocartella in sottocartelle_ordinate:
    percorso_sottocartella = os.path.join(cartella_principale, sottocartella)

    # Chiama la funzione per contare le foto nella sottocartella
    numero_di_foto_sottocartella = conta_foto(percorso_sottocartella)

    # Stampa il nome della sottocartella e il numero di foto
    print(f"Sottocartella: {sottocartella}, Numero di foto: {numero_di_foto_sottocartella}")

    # Aggiungi il numero di foto nella sottocartella al conteggio totale
    totale_foto += numero_di_foto_sottocartella

# Stampa il numero totale di foto in tutte le sottocartelle
print(f"Numero totale di foto in tutte le sottocartelle: {totale_foto}")

```
